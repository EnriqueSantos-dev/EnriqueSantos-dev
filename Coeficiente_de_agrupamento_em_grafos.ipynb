# ***Section - Classes***

---



```python
class MatrizGrafo:


    def __init__(self, vertices):
        self.vertices = vertices
        self.grafo = [[0]*self.vertices for i in range(self.vertices)]

    def adiciona_aresta(self, u, v):
        self.grafo[u-1][v-1] = 1 

    def mostra_matriz(self):
        matriz = []
        for i in range(self.vertices):
            print(self.grafo[i])
```

```python
class ListAd:
    def __init__(self, vertices):
        self.vertices = vertices
        self.grafo = [[] for i in range(self.vertices)]

    def adiciona_aresta(self, u, v):
        self.grafo[u-1].append(v)

    def mostra_lista(self):
        for i in range(self.vertices):
            print(f'{i+1}:', end='  ')
            for j in self.grafo[i]:
                print(f'{j}  ->', end='  ')
            print('')
```

```python
import random
def average_clustering(G, trials=1000):
    n = len(G)
    triangles = 0
    nodes = G.nodes()
    for i in [int(random.random() * n) for i in range(trials)]:
        nbrs = list(G[nodes[i]])
        if len(nbrs) < 2:
            continue
        u, v = random.sample(nbrs, 2)
        if u in G[v]:
            triangles += 1
    return triangles / float(trials)
```

# ***Section - Main***

---



```python
listaDeArestas = []
numeroDeVertices = int(input('Digite a quantidade de vertice do grafo: '))
quantidadeDeArestas = int(input('Quntas aresta vai ter?: '))
for i in  range (quantidadeDeArestas):
    x =  int(input('De qual vertice parte a aresta?: '))
    y =  int (input('Para qual vertice a aresta deve chegar? : '))
    listaDeArestas.append((x,y))
```

# ***Section - Matriz de adjacêcnia e lista de adjacência***

---



```python

gMatriz = MatrizGrafo(numeroDeVertices)
gList = ListAd(numeroDeVertices)
for i in listaDeArestas:
    gMatriz.adiciona_aresta((i[0]), i[1])
    gList.adiciona_aresta((i[0]), i[1])

print('Matriz de adjacência: ')
gMatriz.mostra_matriz()
print('\nLista de adjacência: ')
gList.mostra_lista()
```

# ***Section - Gerando o grafo***

---



###  Gerando um grafo com a quantidade de vértices que foi passada e com as mesmas arestas.





> Criando um grafo com os dados fornecidos pelo usuário na section main a partir da biblioteca networks



```python
import networkx as nx

grafo = nx.Graph()
grafo.add_node(numeroDeVertices)
grafo.add_edges_from(listaDeArestas)
```



> Calculando o coeficiente de agrupamento dos nós do grafo



```python
#Calculando o coeficiente
import networkx as nx
coeficienteDeAgrupamento = nx.average_clustering(grafo)
print(coeficienteDeAgrupamento)
```

### Desenhando o grafo

```python
nx.draw(grafo, with_labels=True)
```
